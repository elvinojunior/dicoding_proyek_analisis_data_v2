### Cleaning Data

```python
linear_cols = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3", "TEMP", "PRES", "DEWP"]

aotiz_df[linear_cols] = aotiz_df[linear_cols].interpolate(method='linear', limit_direction='both')
changping_df[linear_cols] = changping_df[linear_cols].interpolate(method='linear', limit_direction='both')
dingling_df[linear_cols] = dingling_df[linear_cols].interpolate(method='linear', limit_direction='both')
dongsi_df[linear_cols] = dongsi_df[linear_cols].interpolate(method='linear', limit_direction='both')
guanyu_df[linear_cols] = guanyu_df[linear_cols].interpolate(method='linear', limit_direction='both')
gucheng_df[linear_cols] = gucheng_df[linear_cols].interpolate(method='linear', limit_direction='both')
huairo_df[linear_cols] = huairo_df[linear_cols].interpolate(method='linear', limit_direction='both')
nongzhang_df[linear_cols] = nongzhang_df[linear_cols].interpolate(method='linear', limit_direction='both')
shunyi_df[linear_cols] = shunyi_df[linear_cols].interpolate(method='linear', limit_direction='both')
tiantan_df[linear_cols] = tiantan_df[linear_cols].interpolate(method='linear', limit_direction='both')
wanliu_df[linear_cols] = wanliu_df[linear_cols].interpolate(method='linear', limit_direction='both')
wansho_df[linear_cols] = wansho_df[linear_cols].interpolate(method='linear', limit_direction='both')
```

**Insight** :


*   Disini saya menggunakan teknik interpolasi linear untuk pengisian pada kolom ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3", "TEMP", "PRES", "DEWP"] karna polutan udara dan variabel meteorologi umumnya memiliki tren linier dalam jangka pendek
*   Metode ini mempertahankan pola alami perubahan polutan dan cuaca tanpa memperkenalkan bias besar



```python
def fill_missing_values(df):
    """Fungsi untuk mengisi nilai yang hilang sesuai kriteria."""
    if "wd" in df.columns:
        df.loc[:, "wd"] = df["wd"].fillna(df["wd"].mode()[0])

    if "WSPM" in df.columns:
        df.loc[:, "WSPM"] = df["WSPM"].fillna(df["WSPM"].median())

    if "RAIN" in df.columns:
        df.loc[:, "RAIN"] = df["RAIN"].fillna(0)

    return df
```

```python
dataframes = [aotiz_df, changping_df, dingling_df, dongsi_df, guanyu_df,
              gucheng_df, huairo_df, nongzhang_df, shunyi_df, tiantan_df,
              wanliu_df, wansho_df]

dataframes = [fill_missing_values(df) for df in dataframes]
```

**Insight** :

Disini saya membuat fungsi untuk melakukan teknik pengisian missing value menggunakan metode imputation yaitu :
* pada kolom "wd" mengambil nilai modus (nilai paling sering muncul)
* pada kolom "WSPM" mengambil nilai median (nilai tengah)

* pada kolom "RAIN" mengisi dengan nilai 0 (anggap tidak hujan)




```python
def to_datetime(df) :
  df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
  df.drop(columns=['year', 'month', 'day', 'hour'], inplace=True)
  return df

dataframes = [to_datetime(df) for df in dataframes]
```

**Insight** :

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
combined_df["Rain Condition"] = combined_df["RAIN"].apply(lambda x: "No Rain" if x == 0 else "Rain")
grouped_df = combined_df.groupby(["station", "Rain Condition"])["PM2.5"].mean().reset_index()

# Membuat visualisasi menggunakan bar chart
custom_palette = {"No Rain": "orange", "Rain": "skyblue"}
plt.figure(figsize=(12, 6))
sns.barplot(x="station", y="PM2.5", hue="Rain Condition", data=grouped_df, palette=custom_palette)

plt.ylim(0, 120)
plt.yticks(range(0, 121, 10))

# Menambahkan garis horizontal untuk setiap tick di latar belakang
for y in range(0, 121, 10):
    plt.axhline(y=y, color='gray', linestyle='--', alpha=0.5, zorder=0)

plt.xlabel("Stasiun")
plt.ylabel("Rata-rata Konsentrasi PM2.5 (µg/m³)")
plt.title("Perbandingan Rata-rata PM2.5 Saat Hujan dan Tidak Hujan di Setiap Stasiun")
plt.xticks(rotation=45)
plt.legend(title="Kondisi Hujan")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insight** :

* Disini saya melakukan visualisasi menggunakan bar chart agar memudahkan saya dan audiens non-teknis dalam melihat perbandingan antar data 
* Saya membandingkan antara pada saat hujan dengan tidak hujan berdasarkan rata-rata dari PM2.5(polusi partikel 2.5) bisa dilihat dengan adanya hujan (biru) dapat mengurangi polusi partikel 2.5
* kita ambil contoh stasiun Dongsi dengan grafik tertinggi, sebelum hujan rata-rata dari PM2.5 yang diproduksi hingga mendekati 90 namun setelah hujan PM2.5 mereda hingga 60-an

```python
combined_df["Rain Condition"] = combined_df["RAIN"].apply(lambda x: "No Rain" if x == 0 else "Rain")
grouped_df = combined_df.groupby(["station", "Rain Condition"])["PM10"].mean().reset_index()

# Membuat visualisasi menggunakan bar chart
custom_palette = {"No Rain": "orange", "Rain": "skyblue"}
plt.figure(figsize=(12, 6))
sns.barplot(x="station", y="PM10", hue="Rain Condition", data=grouped_df, palette=custom_palette)

plt.ylim(0, 150)
plt.yticks(range(0, 151, 10))

# Menambahkan garis horizontal untuk setiap tick di latar belakang
for y in range(0, 151, 10):
    plt.axhline(y=y, color='gray', linestyle='--', alpha=0.5, zorder=0)

plt.xlabel("Stasiun")
plt.ylabel("Rata-rata Konsentrasi PM10 (µg/m³)")
plt.title("Perbandingan Rata-rata PM10 Saat Hujan dan Tidak Hujan di Setiap Stasiun")
plt.xticks(rotation=45)
plt.legend(title="Kondisi Hujan")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insight** :
* Selanjutnya saya membandingkan antara pada saat hujan dengan tidak hujan berdasarkan rata-rata dari PM10(polusi partikel 10) bisa dilihat dengan adanya hujan (biru) dapat mengurangi polusi partikel 10
* Sebagai contoh yaitu Gucheng dengan rata-rata menghasilkan PM10 sebelum hujan diangka 120, dan setelah hujan menjadi turun hingga sekitar 70

### Pertanyaan 2:

```python
combined_df["year"] = combined_df["datetime"].dt.year

# Menghitung rata-rata PM2.5 dan PM10 per tahun untuk setiap stasiun
pm_yearly_avg = combined_df.groupby(["station", "year"])[["PM2.5", "PM10"]].mean().reset_index()

# Membuat visualisasi tren PM2.5 dan PM10 di setiap stasiun dari tahun ke tahun
fig, axes = plt.subplots(nrows=len(pm_yearly_avg["station"].unique()), figsize=(12, 5 * len(pm_yearly_avg["station"].unique())))

stations = pm_yearly_avg["station"].unique()

for i, station in enumerate(stations):
    subset = pm_yearly_avg[pm_yearly_avg["station"] == station]

    ax = axes[i] if len(stations) > 1 else axes
    ax.plot(subset["year"], subset["PM2.5"], marker="o", linestyle="-", label="PM2.5", color="red")
    ax.plot(subset["year"], subset["PM10"], marker="s", linestyle="--", label="PM10", color="orange")

    ax.set_title(f"Tren Rata-rata Polusi Udara di {station} (PM2.5 & PM10)")
    ax.set_xlabel("Tahun")
    ax.set_ylabel("Konsentrasi (µg/m³)")
    ax.legend()
    ax.grid(True)

    # Menyesuaikan tampilan sumbu x berdasarkan rentang data yang ada
    ax.set_xticks(sorted(subset["year"].unique()))
    ax.set_xlim(subset["year"].min(), subset["year"].max())

    # Menyesuaikan batas sumbu y berdasarkan data yang ada
    ax.set_ylim(0, max(subset[["PM2.5", "PM10"]].max().max(), 150))

plt.suptitle("Perkembangan Kualitas Udara di Setiap Stasiun dari Tahun ke Tahun", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

**Insight:**
- Disini saya menggunakan line chart agar memudahkan saya membandingkan rata-rata pergerakan setiap tahun dari polusi udara yang dihasilkan setiap tahunnya
- Berdasarkan visualisasi data pada tahun 2013-2017 terhadap stasiun yang menghasilkan polusi udara tertinggi dan secara rata-rata di pegang oleh Dongsi dan Gucheng
